### In-Game Purchase Data Analysis

#### Background 

In this homework, you are ask to deal with a task of analyzing an "in-game purchase" dataset. In this game, players are able to to purchase optional items that enhance their playing experience. Now your task is to generate a report that breaks down the game's purchasing data into meaningful insights.

-----

#### Observable Trends

* There are 1163 active players. The vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.79%) with secondary groups falling between 15-19 (18.58%) and 25-29 (13.37%). 

* The age group that spends the most money is the 20-24 with 1,114.06 dollars as total purchase value and an average purchase of 4.32. In contrast, the demographic group that has the highest average purchase is the 35-39 with 4.76 and a total purchase value of 147.67. 
-----

## Purchasing Analysis (Age)

* The below each broken into groups of 4 years (i.e. &lt;10, 10-14, 15-19, etc.)
  * Purchase Count
  * Average Purchase Price
  * Total Purchase Value
  * Average Purchase Total per Person by Age Group
* Bin the purchase_data data frame by age
* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
* Create a summary data frame to hold the results
* Display the summary data frame

```python
#透過定義函數來使呈現更加方便
AgeRange=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
def PC(a,b):
    return len(purchase_data[(purchase_data['Age']>=a) & (purchase_data['Age']<=b)])
def TPV(a,b):
    temp=purchase_data[(purchase_data['Age']>=a) & (purchase_data['Age']<=b)]
    return temp['Price'].sum()
df = pd.DataFrame(
    {
        "Purchase Count": pd.Series([PC(0,9),PC(10,14),PC(15,19),PC(20,24),PC(25,29),PC(30,34),PC(35,39),PC(40,99)],index=AgeRange),
        "Average Purchase Price":pd.Series([TPV(0,9)/PC(0,9),TPV(10,14)/PC(10,14),TPV(15,19)/PC(15,19),TPV(20,24)/PC(20,24),TPV(25,29)/PC(25,29),TPV(30,34)/PC(30,34),TPV(35,39)/PC(35,39),TPV(40,99)/PC(40,99)],index=AgeRange),
        "Total Purchase Value": pd.Series([TPV(0,9),TPV(10,14),TPV(15,19),TPV(20,24),TPV(25,29),TPV(30,34),TPV(35,39),TPV(40,99)],index=AgeRange),
        "Average Purchase Total per Person":pd.Series([TPV(0,9)/Ans(0,9),TPV(10,14)/Ans(10,14),TPV(15,19)/Ans(15,19),TPV(20,24)/Ans(20,24),TPV(25,29)/Ans(25,29),TPV(30,34)/Ans(30,34),TPV(35,39)/Ans(35,39),TPV(40,99)/Ans(40,99)],index=AgeRange)
    }
)
df
```

## Top Spenders

* Identify the the top 5 spenders in the game by total purchase value (top-5 players with highest total purchase value), then list (in a table):
  * SN(screen name)
  * Purchase Count
  * Average Purchase Price
  * Total Purchase Value
* Run basic calculations to obtain the results in the table below
* Create a summary data frame to hold the results
* Sort the total purchase value column in descending order
* Display a preview of the summary data frame

```python
#用groups
IndiPrice=[]
for i in range(len(purchase_data['SN'])):
    Name=purchase_data['SN'][i]
    temp=purchase_data.groupby('SN').get_group(Name)
    IndiPrice.append([temp['Price'].sum(),Name,len(purchase_data.groupby('SN').get_group(Name))])
IndiPrice=sorted(IndiPrice, key = lambda s: s[0], reverse=True)
#print(IndiPrice)//前五依序是[0][5][9][12][16]
TopFiveName=[IndiPrice[0][1],IndiPrice[5][1],IndiPrice[9][1],IndiPrice[12][1],IndiPrice[16][1]]
TopFiveCount=[IndiPrice[0][2],IndiPrice[5][2],IndiPrice[9][2],IndiPrice[12][2],IndiPrice[16][2]]
TopFiveValue=[IndiPrice[0][0],IndiPrice[5][0],IndiPrice[9][0],IndiPrice[12][0],IndiPrice[16][0]]
TopFiveAVG=[IndiPrice[0][0]/IndiPrice[0][2],IndiPrice[5][0]/IndiPrice[5][2],IndiPrice[9][0]/IndiPrice[9][2],IndiPrice[12][0]/IndiPrice[12][2],IndiPrice[16][0]/IndiPrice[16][2]]
df = pd.DataFrame(
    {
        'Purchase Count': pd.Series(TopFiveCount,index=TopFiveName),
        'Average Purchase Price':pd.Series(TopFiveAVG,index=TopFiveName),
        'Total Purchase Value': pd.Series(TopFiveValue,index=TopFiveName)
    }
)
df
```

## Most Popular Items

* Top 5 most popular items by purchase count:
  * Item ID
  * Item Name
  * Purchase Count
  * Item Price
  * Total Purchase Value
* Retrieve the Item ID, Item Name, and Item Price columns
* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
* Create a summary data frame to hold the results
* Sort the purchase count column in descending order
* Display a preview of the summary data frame

```python
#KEY: 不同ID=不同商品
IndiItem=[]
for i in range(len(purchase_data['Item Name'])):
    Name=purchase_data['Item Name'][i]
    ID=purchase_data['Item ID'][i]
    temp=purchase_data.groupby('Item ID').get_group(ID)
    IndiItem.append([temp['Price'].sum(),Name,len(purchase_data.groupby('Item ID').get_group(ID)),ID])
IndiItem=sorted(IndiItem, key = lambda s: s[2], reverse=True)
#print(IndiItem)//前五依序是[0][38][12][15][56]
TopFiveID=[IndiItem[0][3],IndiItem[38][3],IndiItem[12][3],IndiItem[15][3],IndiItem[56][3]]
TopName=[IndiItem[0][1],IndiItem[38][1],IndiItem[12][1],IndiItem[15][1],IndiItem[56][1]]
TopCount=[IndiItem[0][2],IndiItem[38][2],IndiItem[12][2],IndiItem[15][2],IndiItem[56][2]]
TopPrice=[IndiItem[0][0]/IndiItem[0][2],IndiItem[38][0]/IndiItem[38][2],IndiItem[12][0]/IndiItem[12][2],IndiItem[15][0]/IndiItem[15][2],IndiItem[56][0]/IndiItem[56][2]]
TopValue=[IndiItem[0][0],IndiItem[38][0],IndiItem[12][0],IndiItem[15][0],IndiItem[56][0]]
df = pd.DataFrame(
    {
        'Item Name': pd.Series(TopName,index=TopFiveID),
        'Purchase Count': pd.Series(TopCount,index=TopFiveID),
        'Item Price':pd.Series(TopPrice,index=TopFiveID),
        'Total Purchase Value': pd.Series(TopValue,index=TopFiveID)
    }
)
df
```
